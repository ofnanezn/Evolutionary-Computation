## Selection Methods

```python
def roulette_wheel_selection(pop, num_parents):
    selected_parents = []
    total_fitness = sum([ind.fitness for ind in pop])
    for i in range(num_parents):
        r = np.random.uniform()
        p_acum = 0
        for ind in pop:
            p_ind = ind.fitness / total_fitness
            p_acum += p_ind
            if r <= p_acum:
                selected_parents.append(ind)
    return selected_parents
```

```python
def rank_selection(pop, num_parents):
    n = len(pop)
    selected_parents = []
    ranked_pop = sorted(pop, key=lambda x: x.fitness)
    p = [ranked_pop[i].fitness / (n*(n-1)) for i in range(len(ranked_pop))]
    for i in range(num_parents):
        r = np.random.uniform()
        p_acum = 0
        for i in range(len(ranked_pop)):
            p_acum += p[i]
            if r <= p_acum:
                selected_parents.append(ranked_pop[i])
    return selected_parents
```

## Replacement Methods

```python
def generational_replacement(parents, children):
    return children
```

```python
def stable_state_replacement(parents, children):
    return sorted(parents + children, key=lambda x: x.fitness, reverse=True)[:len(parents)]
```

```python
def GeneticAlgorithmIteration(fitness, T, chrom_size, P):
    GA = GeneticAlgorithm(chrom_size)
    pop = [GA.create_individual() for i in range(P)]
    for individual in pop:
        individual.calculate_fitness(fitness)
    for t in range(T):
        parents = rank_selection(pop, num_parents)
        #print(parents)
        new_pop = []
        for i in range(int(P/2)):
            parent1 = parents[np.random.randint(0, len(parents))]
            parent2 = parents[np.random.randint(0, len(parents))]
            cross_point = np.random.randint(1,chrom_size-1)
            child1, child2 = GA.crossover(parent1.chromosome, parent2.chromosome, cross_point)
            child1 = GA.mutation(child1.chromosome)
            child2 = GA.mutation(child2.chromosome)
            child1.calculate_fitness(fitness)
            child2.calculate_fitness(fitness)
            new_gen = stable_state_replacement([parent1, parent2],[child1, child2])
            for ind in new_gen:
                new_pop.append(ind)
            print(max(pop, key=lambda x: x.fitness).fitness)
        pop = new_pop
    return max(pop, key=lambda x: x.fitness)
```

```python
def MaxOnes_fitness(chromosome):
    cnt = 0
    while(chromosome):
        cnt += chromosome & 1
        chromosome >>= 1
    return cnt
```

```python
best = GeneticAlgorithmIteration(MaxOnes_fitness, 100, 99, 100)
```
