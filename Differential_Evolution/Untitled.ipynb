```python
import numpy as np
from copy import deepcopy
```

```python
def pick_dif_parents(P, i):
    gen = [i]
    while len(gen) < 4:
        r = np.random.randint(len(P))
        if r not in gen:
            gen.append(r)
    return gen[1], gen[2], gen[3]
```

```python
def next_ind_de(P, i):
    (a, b, c) = pick_dif_parents(P, i)
    d = len(P[i])
    R = np.random.randint(0, d)
    q = np.zeros(d)
    for k in range(d):
        if np.random.uniform <= 0.5 or k == R:
            q[k] = P[a][k] + 0.5 * (P[b][k] - P[c][k])
        else:
            q[k] = P[i][k]
    return q
```

```python
def next_pop_de(P, f=lambda x: x):
    Q = deepcopy(P)
    for i in range(len(P)):
        child = next_ind_de(P, i)
        if f(child) <= f(Q[i]):
            Q[i] = child
    return Q
```

```python
def DE(T, fitness):
    P = [np.random.uniform(-5, 5, size=2) for i in range(100)]
    print([f(i) for i in P])
    for i in range(T):
        P = next_pop_de(P, fitness)
    return P
```

```python
def f(x):
    return np.sum(x ** 2) + 5
```

```python
P = DE(100, f)
```

```python
print([f(i) for i in P])
```
