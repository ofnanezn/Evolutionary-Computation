# Neuro-Evolution

## Extracting the Dataset

```python
def shuffle_dataset(X, Y):
    p = np.random.permutation(X.shape[1])
    return X[:,p], Y[:,p]
```

```python
data = load_iris()
percent_train = 0.8
num_examples = data.data.shape[0]
num_dim = data.data.shape[1]
num_classes = 3

X = data.data.T
labels = data.target
Y = np.eye(num_classes)[labels].T

X_shuffled, Y_shuffled = shuffle_dataset(X, Y)
```

```python
X_train = X_shuffled[:, :int(num_examples*percent_train)]
X_test = X_shuffled[:, int(num_examples*percent_train):]

Y_train = Y_shuffled[:, :int(num_examples*percent_train)]
Y_test = Y_shuffled[:, int(num_examples*percent_train):]
```

## Creating the Neural Network

```python
hidden_layers = [5, 4, 4]
activation_functions = ["RELU", "RELU", "RELU", "softmax"]
    
NN = NeuralNetwork(hidden_layers, activation_functions)
```

```python
W, b = NN.initialize_weights(X_train.shape[0], Y_train.shape[0])
```

```python
for l in range(len(b)):
    print(b[l].shape)
```

## Testing the Neural Network Functionalities

## Experiments

### Hill-Climbing:

```python
def standarDeviation(x):
    return np.sqrt(np.sum((x-np.median(x))**2)/len(x))
```

```python
def getErrorPlot(data, num_steps=100):
    x = list(range(0, data.shape[1], num_steps))
    y = np.empty(0)
    std = np.empty(0)
    for j in range(0, data.shape[1], num_steps):
        xj = data[:, j]
        y = np.append(y, np.median(xj))
        std = np.append(std, standarDeviation(xj))
    return x, y, std
```

```python
def plotGraphs(graph):
    plt.plot(list(range(1,10001)), np.average(graph, axis=0), label='Average')
    plt.plot(list(range(1,10001)), np.median(graph, axis=0), label='Median')
    plt.plot(list(range(1,10001)), np.min(graph, axis=0), label='Min')
    plt.plot(list(range(1,10001)), np.max(graph, axis=0), label='Max')
    x, y, z = getErrorPlot(graph)
    plt.errorbar(x, y, z, marker='^', ls='None')
    plt.legend(loc='upper right', shadow=True, fontsize = "xx-small")
    plt.xlabel('iterations')
    plt.ylabel('f(x)')
```

```python
def hill_climbing_experiments(X, y, num_experiments=30, powerLaw=True, delta=0.1):
    scores = []
    total_history = []
    for i in range(num_experiments):
        X_shuffled, Y_shuffled = shuffle_dataset(X, y)
        X_train = X_shuffled[:, :int(num_examples*percent_train)]
        X_test = X_shuffled[:, int(num_examples*percent_train):]

        Y_train = Y_shuffled[:, :int(num_examples*percent_train)]
        Y_test = Y_shuffled[:, int(num_examples*percent_train):]

        hidden_layers = [5, 4, 4]
        activation_functions = ["RELU", "RELU", "RELU", "softmax"]

        NN = NeuralNetwork(hidden_layers, activation_functions)
        W, b = NN.initialize_weights(X_train.shape[0], Y_train.shape[0])
        parameters, history_loss = NN.HillClimbingTrain(X_train, W, b, Y_train, delta=delta, num_iterations=10000)
        score = NN.evaluateModel(X_test, Y_test, parameters)
        scores.append(score)
        total_history.append(history_loss)
    return scores, np.array(total_history)
```

```python
scores_hill_climbing, total_history = hill_climbing_experiments(X, Y, delta=0.01)
```

```python
N = len(scores_hill_climbing)
mean_HC = sum(scores_hill_climbing) / N
std_dev_HC = sum([(scores_hill_climbing[i] - mean_HC)**2 for i in range(N)])/N
```

```python
print("Mean:", mean_HC)
print("Standard Deviation:", std_dev_HC)
```

```python
plt.figure(figsize=(10,8))
plotGraphs(total_history)
```

### Simulated Annealing

```python
def simulated_annealing_experiments(X, y, num_experiments=30, powerLaw=True, delta=0.1):
    scores = []
    total_history = []
    for i in range(num_experiments):
        X_shuffled, Y_shuffled = shuffle_dataset(X, y)
        X_train = X_shuffled[:, :int(num_examples*percent_train)]
        X_test = X_shuffled[:, int(num_examples*percent_train):]

        Y_train = Y_shuffled[:, :int(num_examples*percent_train)]
        Y_test = Y_shuffled[:, int(num_examples*percent_train):]
        
        hidden_layers = [5, 4, 4]
        activation_functions = ["RELU", "RELU", "RELU", "softmax"]

        NN = NeuralNetwork(hidden_layers, activation_functions)
        W, b = NN.initialize_weights(X_train.shape[0], Y_train.shape[0])
        parameters, history_loss = NN.SimulatedAnnealingTrain(X_train, W, b, Y_train, delta=delta, t_max=10000, powerLaw=powerLaw)
        total_history.append(history_loss)
        score = NN.evaluateModel(X_test, Y_test, parameters)
        scores.append(score)
    return scores, np.array(total_history)
```

```python
scores_simulated_annealing, total_history = simulated_annealing_experiments(X, Y, delta=0.01, powerLaw=False)
```

```python
N = len(scores_simulated_annealing)
mean_SA = sum(scores_simulated_annealing) / N
std_dev_SA = sum([(scores_simulated_annealing[i] - mean_SA)**2 for i in range(N)])/N
```

```python
print("Mean:", mean_SA)
print("Standard Deviation:", std_dev_SA)
```

```python
plt.figure(figsize=(10,8))
plotGraphs(total_history)
```
